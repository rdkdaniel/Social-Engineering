```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
legit_urls = pd.read_csv('legitimate-urls.csv')
phish_urls = pd.read_csv('phishing-urls.csv')
```

```python
legit_urls.head(11)
```

```python
phish_urls.head(11)
```

```python
#Drop columns not needed
phish_urls = phish_urls.drop(phish_urls.columns[[0,3,5]],axis=1)
```

```python
phish_urls.head(1)
```

```python
legit_urls = legit_urls.drop(legit_urls.columns[[0,3,5]],axis=1)
```

```python
legit_urls.head(1)
```

# **Combining these two dataset to 1**

```python
#using merge function by setting how='outer'
phishdata = pd.merge(legit_urls, phish_urls, 
                   on='Having_@_symbol', 
                   how='outer')
```

```python
phishdata = pd.read_csv ('phishdata.csv')
```

```python
# displaying result
phishdata.head(10)
```

# **Saving new dataset**

```python
phishdata.to_csv('phishdata.csv')
#Went ahead to download from files
```

# **Model**

```python
#Defining the "Features" and "Target" for the model.
#These items form the basis of the train-test-split procedure
#Remember, the features are the attributes that form the basis of the prediction and the target is the output/outcome
features = ['Having_@_symbol','Redirection_//_symbol_x','Sub_domains_x','URL_Length_x','age_domain_x']
data = phishdata.loc[:, features] #features
labels = phishdata.loc[:, ['label_y']] #target
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split the data set into testing and training parts
data_train, data_test, labels_train, labels_test = train_test_split(phishdata, labels, test_size=0.30, random_state=110)
```

```python
#Making model instance
reg = RandomForestClassifier(max_depth = 2, random_state = 0)
```

```python
#training model on data
reg.fit(data_train, labels_train)
```

```python
#The error above is due to Null values - drop them and repeat to continue
```

```python
#Predict labels of unseen data
# Predicting multiple observations
reg.predict(data_test[0:10])
```

```python
data_test.head(1)
```

```python
# predict 1 observation.
reg.predict(data_test.iloc[0].values.reshape(1,-1))
```

```python
#Measure train test split performance
score = reg.score(data_test, labels_test)
print(score)
```
